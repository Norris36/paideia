```python
skip = []


for i in range(len(df)):
    if i not in skip:
        x1 = df.at[i, 'text']
        y1 = df.at[i, 'category']
        
        if y1 == 'Generative Pre-trained Transformers':
            print('hi')
            if i < len(df)-2:
                j = i
                while len(x1) < 2400:
                    j += 1
                    skip.append(j)
                    y2 = df.at[j, 'category']
                    if y1 == y2:
                        x2 = df.at[j, 'text']
                        x1 += ' ' + x2
                    else:
                        break
                df.at[i, 'text'] = x1           
    else:
        df.drop(i, axis = 0, inplace=True)
```
