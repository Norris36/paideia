```python
def mean_getter(ingredient = ingredient, measurement = measurement, prefered_list = salads):
    grouped             = pantry[(pantry.ingredients == ingredient)
                                  & (pantry.units == measurement)
                                  & (pantry.urls.isin(prefered_list))].groupby(['amounts', 'units']).count().reset_index()

    grouped.amounts     = grouped.amounts.astype(float)
    grouped.units       = grouped.units.astype(str)
    grouped.amounts     = grouped.amounts * grouped.urls
    average             = grouped.amounts.sum() / grouped.urls.sum()
    return average

mean_getter()
```
