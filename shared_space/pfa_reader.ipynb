```python
def get_modified_time(path):
    #Gaetting the modified time of the file.
    return os.stat(path).st_mtime

import os
from pathlib import Path

def get_latest_folder(directory_path):
    """
    Finds the most recently created folder in a given directory.

    Args:
    directory_path (str): The path of the directory to search in.

    Returns:
    str: The full path of the most recently created folder.
    """
    # Convert the string path to a Path object for easy manipulation
    directory = Path(directory_path)

    # Initialize variables to store the latest folder and its creation time
    latest_folder = None
    latest_time = 0

    # Iterate through each item in the directory
    for item in directory.iterdir():
        # Check if the item is a directory
        if item.is_dir():
            # Get the creation time of the directory
            creation_time = os.path.getctime(item)

            # If this directory is newer than the current 'latest', update 'latest_folder' and 'latest_time'
            if creation_time > latest_time:
                latest_folder = item
                latest_time = creation_time

    # Return the path of the latest folder, or None if no folders were found
    return str(latest_folder) if latest_folder else None
latest_folder_path = get_latest_folder(r"C:\Users\[user]\Documents\ShareX\Screenshots")

# Example usage
print("Latest folder:", latest_folder_path)


def remove_waste_in_downloads():
    #Removing files in the downloads folder that are older than 8 days.
    down = my.get_downloads()

    #Getting a set of paths from the downloads folder
    paths = my.hygin(down, '')



    #Setting the time barrier, the cut off point, so the current time minus 8 days (24 hours consisting of 60 minutes consisting of 60 seconds)
    time_barrier = datetime.timestamp(datetime.utcnow()) - 8*((24 * (60 * 60)))
    time_barrier = datetime.timestamp(datetime.utcnow()) - 8*((24 * (60 * 60)))
    barier_date = datetime.fromtimestamp(time_barrier).strftime("%x")
    barier_time = datetime.fromtimestamp(time_barrier).strftime("%X")
    print("the deletion date for files is ",barier_date, barier_time)

    #Iterating through the paths, and checking the modified time against the time barrier
    for path in paths:
        if 'Downloads' in path:
            if get_modified_time(path) < time_barrier:
                #actually deleting files
                os.remove(path)
            
            
    # lets walk thorugh the downloads folder and get all the folders
    # then we will check if the folder is older than 8 days
    # if it is older than 8 days, we will delete it or if they are empty
    for root, dirs, files in os.walk(down):
        for name in dirs:
            if 'Downloads' in path:
            #print(os.path.join(root, name))
                if get_modified_time(os.path.join(root, name)) < time_barrier:
                    #actually deleting files
                    os.rmdir(os.path.join(root, name))
                    print("deleted folder: ", os.path.join(root, name))
                elif len(os.listdir(os.path.join(root, name))) == 0:
                    os.rmdir(os.path.join(root, name))
                    print("deleted folder: ", os.path.join(root, name))
                
remove_waste_in_downloads()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

try:
    if new_df is None:
        new_df = pd.read_csv('pfa.csv')
    else:
        print('new_df already exists')
        pass
except Exception as e:
    print(e)
    print(type(e))
    new_df = pd.read_csv('pfa.csv')

my_dict = {}

for col in new_df.columns:
    if col != 'timestamp' and col != 'date' and col != 'week' and col != 'Frederiksberg Alle' and col != 'Vodroffsvej' and col != 'Mariendalsvej' and col != 'Nyelandsvej':
        # create an empty dictionary to store the results

        # replace 'col' with the name of the column you want to analyze
        #col = 'my_column_name'

        # filter out rows where the column is null
        filtered_df = new_df[new_df[col].notnull()].copy()

        # convert the column to a numeric data type
        filtered_df[col] = pd.to_numeric(filtered_df[col], errors='coerce')

        # remove rows where the column is NaN
        filtered_df = filtered_df.dropna(subset=[col])

        # perform the linear regression
        x = filtered_df['timestamp']
        y = filtered_df[col]
        a, b = np.polyfit(x, y, 1)

        # find the x where y=0 (the intersection with x-axis)
        x_intersect = -b / a

        # calculate the x-intercept for the subset of data that excludes the most recent data point
        last_week = filtered_df[:len(filtered_df) - 1]
        x1 = last_week['timestamp']
        y1 = last_week[col]
        a1, b1 = np.polyfit(x1, y1, 1)
        x_intersect1 = -b1 / a1
        x_intersect = datetime.fromtimestamp(x_intersect).strftime('%d-%m-%Y')  
        x_intersect1 = datetime.fromtimestamp(x_intersect1).strftime('%d-%m-%Y')
        my_dict[col] = {'a': a, 'b': b, 'x_intersect': x_intersect, 'x_intersect1': x_intersect1}
    
```
